```python
import matplotlib.pyplot as plt
import numpy as np


frequenze = {
    'a': 0.12377353175907234,
    'b': 0.016525045772498945,
    'c': 0.04314938265809117,
    'd': 0.048689028684099335,
    'e': 0.13180719215060327,
    'f': 0.004160602788601474,
    'g': 0.012229472794704474,
    'h': 0.0163842073142106,
    'i': 0.04827238157832966,
    'j': 0.005252100840336135,
    'k': 8.215576733486691E-5,
    'l': 0.05320172761842167,
    'm': 0.03140697619830055,
    'n': 0.06599455424627952,
    'o': 0.10111027651283977,
    'p': 0.03176494061311676,
    'q': 0.01394887563964133,
    'r': 0.0687761137974743,
    's': 0.06756725036383268,
    't': 0.039563870240833764,
    'u': 0.04548495375803953,
    'v': 0.010956058401014037,
    'w': 0.0,  
    'x': 6.807192150603258E-4,
    'y': 0.013761091028590206,
    'z': 0.005457490258673302
}


letters = list(frequenze.keys())
frequencies = list(frequenze.values())

sorted_indices = np.argsort(frequencies)[::-1]  
sorted_letters = [letters[i] for i in sorted_indices]
sorted_frequencies = [frequencies[i] for i in sorted_indices]


cmap = plt.get_cmap('RdYlGn')
norm = plt.Normalize(min(sorted_frequencies), max(sorted_frequencies))
colors = cmap(norm(sorted_frequencies))


plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_letters, sorted_frequencies, color=colors)

plt.xlabel('Letters', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.title('Letter Frequencies in *Language*', fontsize=25)
plt.xticks(rotation=45, fontsize=17)
plt.yticks(fontsize=17)

for y in np.arange(0.02, max(sorted_frequencies)+0.02 , 0.02):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=0.5)


plt.tight_layout()
plt.show()
```
