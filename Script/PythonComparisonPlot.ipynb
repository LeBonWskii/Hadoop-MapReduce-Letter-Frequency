```python
import matplotlib.pyplot as plt

file_sizes = ['2KB', '2MB', '2GB']
cpu_time_mapreduce = [2400, 4490, 756910]  
cpu_time_python = [31.25, 1140.62, 1181515.62]  

# Plot per il CPU time
plt.figure(figsize=(12, 8))
mapreduce_plot, = plt.plot(file_sizes, cpu_time_mapreduce, 'o-', label='MapReduce')
python_plot, = plt.plot(file_sizes, cpu_time_python, 's--', label='Python')
plt.xlabel('File Size')
plt.ylabel('CPU Time (ms)')
plt.title('CPU Time spent (ms)')
plt.legend()
plt.grid(True)

# Aggiungi il valore di ogni punto nel grafico con un offset
offset = max(max(cpu_time_mapreduce), max(cpu_time_python)) * 0.02  # Calcola un offset come percentuale dei valori massimi

for i, txt in enumerate(cpu_time_mapreduce):
    plt.text(file_sizes[i], cpu_time_mapreduce[i] - 1.6*offset, f'{txt}', color=mapreduce_plot.get_color(), ha='center', va='top')
for i, txt in enumerate(cpu_time_python):
    plt.text(file_sizes[i], cpu_time_python[i] + offset, f'{txt}', color=python_plot.get_color(), ha='right', va='bottom')

plt.show()
```
