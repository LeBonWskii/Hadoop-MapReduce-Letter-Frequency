```python
import matplotlib.pyplot as plt


plt.rcParams.update({'font.size': 16})

file_sizes = ['2KB', '2MB', '2GB']
cpu_time_mapreduce = [2400, 4490, 756910]  
cpu_time_python = [31.25, 1140.62, 1181515.62]  


plt.figure(figsize=(20, 14))
mapreduce_plot, = plt.plot(file_sizes, cpu_time_mapreduce, 'o-', label='Hadoop')
python_plot, = plt.plot(file_sizes, cpu_time_python, 's--', label='Python')
plt.xlabel('File Size', fontsize=18)
plt.ylabel('CPU Time (ms)', fontsize=18)
plt.title('CPU Time spent (ms)', fontsize=20)
plt.legend(fontsize=16)
plt.grid(True)


offset = max(max(cpu_time_mapreduce), max(cpu_time_python)) * 0.02  

for i, txt in enumerate(cpu_time_mapreduce):
    plt.text(file_sizes[i], cpu_time_mapreduce[i] - 1.6*offset, f'{txt}', color=mapreduce_plot.get_color(), ha='center', va='top', fontsize=14)
for i, txt in enumerate(cpu_time_python):
    plt.text(file_sizes[i], cpu_time_python[i] + offset, f'{txt}', color=python_plot.get_color(), ha='right', va='bottom', fontsize=14)

plt.show()
```
